```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#jupyter上でグラフを表示できるようにする
%matplotlib inline
```

```python
start = '2019-06-01'
end = '2020-06-01'

#^N225はティッカーシンボル　日経平均
df = data.DataReader('^N225', 'yahoo', start, end)
```

```python
df.head(10)
```

```python
df.index
```

```python
date = df.index
price=df['Adj Close']
```

```python
plt.plot(date, price)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
plt.title('N225',color='blue', backgroundcolor='white', size=40, loc='center')
plt.xlabel('date',color='black',size=30)
plt.ylabel('price', color='black', size=30)
plt.legend()
```

```python
span01 = 5
span02 = 25
span03 = 50

df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()
```

```python
pd.set_option('display.max_rows', None)
df.head(100)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
plt.plot(date,df['sma01'], label='sma01')
plt.plot(date,df['sma02'], label='sma02')
plt.plot(date,df['sma03'], label='sma03')


plt.title('N225',color='blue', backgroundcolor='white', size=40, loc='center')
plt.xlabel('date',color='black',size=30)
plt.ylabel('price', color='black', size=30)
plt.legend()
```

### いつ買えばよかったのか？？

ゴールデンクロス=短期の移動平均線が中期以上の移動平均線を下から上へに抜けること
価格が上昇しているサイン

デッドクロス＝短期の移動平均線が中期以上の移動平均線を上から下に抜けること。価格下降しているサイン

価格が下がるときは空売りで設けることができる

```python
date=df.index
price=df['Close']

span01 = 5
span02 = 25
span03 = 50

df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()

plt.figure(figsize=(30,15))
plt.subplot(2,1,1)

plt.plot(date, price, label='Close', color='#99b898')
plt.plot(date,df['sma01'], label='sma01')
plt.plot(date,df['sma02'], label='sma02')
plt.plot(date,df['sma03'], label='sma03')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df['Volume'],label='Volume', color='grey')
plt.legend()
```
